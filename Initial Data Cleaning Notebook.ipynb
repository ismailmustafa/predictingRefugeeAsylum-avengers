```python
import math

def valid_asy_type(asy_type):
    if isinstance(asy_type, float) and math.isnan(asy_type):
        return True
    if isinstance(asy_type, str) and ('FEB' in asy_type or 'MAR' in asy_type):
        return False
    if asy_type in ['E', 'I', '0']:
        return True
    return True

df2 = df[df['c_asy_type'].map(valid_asy_type)]
df2 = df2[df2['ij_code'] != '1996']
df2.head(10)
```
